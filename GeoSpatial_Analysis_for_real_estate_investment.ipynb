# <span style='color:blue'>  MODELLING THE HOUSE PRICE APPRECIATION </span>

## <span style='color:blue'>   Read data and Drop duplicates from dataset  </span>

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
#plt.style.use('ggplot')

import sys
import warnings

if not sys.warnoptions:
    warnings.simplefilter("ignore")
    
    
def get_y_var(row):
    if row.Year < 2017 : 
        d = data[(data.Year == row.Year+1) & (data.Tract_number == row.Tract_number)]
        if len(d) > 0 : 
            return list(d['Median Estimated Home Value(owner occupied units)'])[0]
        else: # We return -1 when we dont have the future value. 
            return -1
    else:
        return -1 
    
    
#data['y_var'] = data.apply(lambda row : get_y_var(row), axis = 1)
#data.to_csv('data_with_y_var.csv')


def adjust_price(row):
    df = data[data.Tract_number == row.Tract_number] [['Median Estimated Home Value(owner occupied units)', 
      'zip_code', 'Year', 'state', 'Tract_number']]

    price_mean = df['Median Estimated Home Value(owner occupied units)'].mean()
    
    # 40% deviation is allowed 
    lower_price_limit = price_mean * 0.6
    upper_price_limit = price_mean * 1.4
    
    current_price = row['Median Estimated Home Value(owner occupied units)']
    
    if (current_price <= lower_price_limit) or (current_price >= upper_price_limit):
        return price_mean
    else : 
        return current_price
    
    
    

data = pd.read_csv('data_x.csv')
data = data.drop_duplicates()
```

## <span style='color:blue'>   Outlier Treatment   </span>

## <span style='color:blue'>   Missing Value treatment - Median Imputation (For zero and Nan)</span>

- Here we can see that around 16% of the data is filled with zeroes which are missing values. 
- So we impute the nan values and zero values through median imputation method. 
- We start off by imputing by grouping on zip code but turns out that there are scenarios where all the values are zeroes for a specific zip code. 
- In such a case, take a 4 digit zip level and do the grouping (and so on )

```python
d = data[(data["Unweighted Sample Count of the population"] == 0) 
         |(data["Total population in occupied housing units"] == 0) 
         |(data["Median Estimated Home Value(owner occupied units)"] == 0) 
         |(data["Total Population"] == 0) |(data["Total Population over 25"] == 0) 
         |(data["Homeowner households"] == 0) |(data["Renter households"] == 0) 
         |(data["Housing units with mortgage"] == 0) |(data["Total housing Units"] == 0) 
         |(data["Number of Sales and office workers"] == 0) 
         |(data["Number of people in management business science and arts"] == 0) 
         |(data["Number of service workers"] == 0) 
         |(data["Number of educational and health service workers"] == 0) 
         |(data["Number of artsentertainment and food service workers"] == 0) 
         |(data["Number of finance and real estate workers"] == 0) 
         |(data["Number of tech workers"] == 0) ]

len(d)/len(data)
```

```python
#Replacing zeros with nan values so that we can adjust nans with median Imputation 
data[cols_with_high_zeros] = data[cols_with_high_zeros].replace({'0':np.nan, 0:np.nan, '0.0':np.nan, 0.0 : np.nan})
data.isna().sum().sum()
```

```python
#Treatment for nans 
data['zip_code'].fillna(method='ffill', inplace=True)
data['Unweighted Sample Count of the population'].fillna(0,  inplace=True)
data["Median Gross Rent"] = data.groupby("zip_code")["Median Gross Rent"].transform(lambda x: x.fillna(x.median()))
data["Median Household Income"] = data.groupby("zip_code")["Median Household Income"].transform(lambda x: x.fillna(x.median()))


for col in cols_with_high_zeros:
    data[col] = data.groupby("zip_code")[col].transform(lambda x: x.fillna(x.median()))

data.isna().sum().sum()
```

```python
data.zip_code = data.zip_code.astype(int)
data['zip_1'] = data.zip_code//10
data['zip_2'] = data.zip_code//10//10

for col in cols_with_high_zeros:
    data[col] = data.groupby("zip_1")[col].transform(lambda x: x.fillna(x.median()))

    
for col in cols_with_high_zeros:
    data[col] = data.groupby("zip_2")[col].transform(lambda x: x.fillna(x.median()))
data = data.drop(['zip_1', 'zip_2'], axis =1 )
data.to_csv('data_post_missing_val_treatment.csv')

data.isna().sum().sum()
```

- To avoid data leakage, we scale the test data using a scaler fitted by historical values. 
- We also scale the y variables to have a better fit. This is a very important step in regression where we use disance based algorithms for modelling

## <span style='color:blue'>  Simple Linear Regression   </span>

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error 
from sklearn.model_selection import cross_val_score
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import RandomizedSearchCV
from scipy.stats import randint

def display_scores(scores):
    print("Scores:", scores)
    print("Mean:", scores.mean())
    print("Standard deviation:", scores.std())


def evaluate(model, X_test, Y_test):
    predictions = model.predict(X_test)
    errors = abs(predictions - Y_test)
    mape = 100 * np.mean(errors / Y_test)
    accuracy = 100 - mape
    print('Model Performance')
    print('Average Error: {:0.4f} degrees.'.format(np.mean(errors)))
    print('Accuracy = {:0.2f}%.'.format(accuracy))
    
    return accuracy, predictions
```

```python
lin_reg = LinearRegression()
lin_reg.fit(X_train, Y_train)

 
lin_reg_preds = lin_reg.predict(X_test) 
lin_mse = mean_squared_error(Y_test, lin_reg_preds) 
lin_rmse = np.sqrt(lin_mse) 
lin_rmse
```

- High cv of 10, reduces the chance of overfitting
- High n_iter of 50 will cover a wider search space 

```python
# Randomized hyper parameter search


# Number of trees in random forest
n_estimators = [int(x) for x in np.linspace(start = 200, stop = 2000, num = 10)]

#Number of features to consider at every split
max_features = ['auto', 'sqrt']

# Maximum number of levels in tree
max_depth = [int(x) for x in np.linspace(10, 110, num = 11)]
max_depth.append(None)

# Minimum number of samples required to split a node
min_samples_split = [2, 5, 10]

# Minimum number of samples required at each leaf node
min_samples_leaf = [1, 2, 4]


# Create the random grid
random_grid = {'n_estimators': n_estimators,
               'max_features': max_features,
               'max_depth': max_depth,
               'min_samples_split': min_samples_split,
               'min_samples_leaf': min_samples_leaf}


forest_reg = RandomForestRegressor(random_state=42)
rnd_search = RandomizedSearchCV(estimator = forest_reg, param_distributions=param_distribs,
                                n_iter=20, cv=10, scoring='neg_mean_squared_error', random_state=42, verbose =2 )

rnd_search.fit(X_train, Y_train)
cvres = rnd_search.cv_results_
for mean_score, params in zip(cvres["mean_test_score"], cvres["params"]):
    print(np.sqrt(-mean_score), params)
```
